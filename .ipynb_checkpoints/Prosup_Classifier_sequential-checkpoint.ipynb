```python
def filter_df(files_df,order=1,cf=50,fs=2000):
    emg_labels = ['EMG1','EMG2','EMG3','EMG4','EMG5','EMG6','EMG7','EMG8']
    angle_lables = ['Pronation_Angle','Flexion_Angle',  'Radial_Angle','Elbow_Joint_Angle']
    b,a = signal.butter(order, cf,fs=fs)
    for i in range(len(files_df)):
        emg_df = pd.DataFrame(columns=emg_labels)
        for labels in emg_labels:
            emg_df[labels] = signal.lfilter(b, a,files_df[i][labels])
        files_df[i][emg_labels] = np.array(emg_df)
        
        b,a = signal.butter(3, 3,fs=2000)
        for labels in angle_lables:
            files_df[i][labels] = signal.lfilter(b, a,files_df[i][labels])
            
    return files_df
```

### TW Normalize

time window normalization (i.e., subtracting to
each time window the mean and dividing it by the standard
deviation) and normalization based on training data (i.e.,
subtracting to all the time windows the training data mean and
dividing them by the training data standard deviation).

```python
def variance(data, ddof=0):
    n = len(data)
    mean = sum(data) / n
    return sum((x - mean) ** 2 for x in data) / (n - ddof)
```

```python
X_mean = np.zeros(X_prosup[0].shape)
for x in X_prosup:
    X_mean = X_mean + x
X_mean = X_mean/X_prosup.shape[0]
X_std = np.sqrt(variance(X_prosup))
np.save('mean_std',np.array([X_mean,X_std]))
```

# InceptionTime Trainer

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import LSTM
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical
from tensorflow import keras
n_outputs = 3
```

```python
checkpoint_filepath = 'best_model_seq_prosup.hdf5'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
n_outputs = 3
filter_size = 11
in_shape = (X_train.shape[1],X_train.shape[2])

model = Sequential()

model.add(Conv1D(filters=128, kernel_size=filter_size, activation='relu', input_shape=in_shape))
model.add(Conv1D(filters=128, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=128, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=128, kernel_size=filter_size, activation='relu'))

model.add(Conv1D(filters=64, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=filter_size, activation='relu'))

model.add(Dropout(0.5))

model.add(Conv1D(filters=32, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=32, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=32, kernel_size=filter_size, activation='relu'))
model.add(Conv1D(filters=32, kernel_size=filter_size, activation='relu'))

# model.add(Dropout(0.5))

# model.add(Conv1D(filters=16, kernel_size=filter_size, activation='relu'))
# model.add(Conv1D(filters=16, kernel_size=filter_size, activation='relu'))
# model.add(Conv1D(filters=16, kernel_size=filter_size, activation='relu'))
# model.add(Conv1D(filters=16, kernel_size=filter_size, activation='relu'))

# model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Flatten())
model.add(Dense(16,activation='relu'))
# model.add(Dense(8,activation='relu'))

model.add(Dense(n_outputs,activation='softmax'))

# model.compile(loss='mean_squared_error', optimizer='adam')
# model.compile(loss='mean_squared_error', optimizer='adam',metrics=[pearson_r])

model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])



model.summary()
```

```python
cnn_history=model.fit(X_train,y_train,epochs = 100,batch_size = 512,validation_data = (X_test,y_test),callbacks=[model_checkpoint_callback])
```

```python
best_model = load_model('best_model_seq_prosup.hdf5')
best_model.save('models/prosup_seq_{}.hdf5'.format(250))
```

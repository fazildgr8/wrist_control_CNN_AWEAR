# Test Data Prep and Deep Learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from time import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
file = 'Subjects/Fazil/day_04_24_21/computed_Hammer.csv'
df = pd.read_csv(file,index_col=0)
```

```python
def prep_data(df,window,angle_label,Normalize=True,file_save=False):
    
    angle_thresh = 0.01
    
    emg_df = df[['EMG1', 'EMG2', 'EMG3', 'EMG4', 'EMG5', 'EMG6',
           'EMG7', 'EMG8', 'EMG9', 'EMG10', 'EMG11', 'EMG12']]
    scaler = StandardScaler(with_mean=True,
                            with_std=True,
                            copy=False).fit(emg_df)
    if(Normalize==True):
        emg_df = scaler.transform(emg_df)
    emg_array = np.array(emg_df)
    all_angle = np.array(df[angle_label])
    X = []
    y = []
    for i in range(df.shape[0]-window):
        j = i + window
        loc_arr = emg_array[j-window:j]
        X.append(loc_arr)
        angles = all_angle[j-window:j]
        diff = difference(angles).mean()
        if(abs(diff)<angle_thresh):
            y.append([1,0,0])
        elif(diff>0):
            y.append([0,1,0])
        else:
            y.append([0,0,1])
    X = np.array(X)
    y = np.array(y)
        
    return X, y

def difference(dataset, interval=1):
	diff = []
	for i in range(interval, len(dataset)):
		value = dataset[i] - dataset[i - interval]
		diff.append(value)
	return np.array(diff)
```

```python
def prep_data_DTM(df,window,Normalize=True,file_save=False):
    
    angle_thresh = 0.008
    
    emg_df = df[['EMG1', 'EMG2', 'EMG3', 'EMG4', 'EMG5', 'EMG6',
           'EMG7', 'EMG8', 'EMG9', 'EMG10', 'EMG11', 'EMG12']]
    scaler = StandardScaler(with_mean=True,
                            with_std=True,
                            copy=False).fit(emg_df)
    if(Normalize==True):
        emg_df = scaler.transform(emg_df)
    emg_array = np.array(emg_df)
    all_angle_flexion = np.array(df['Flexion_Angle'])
    all_angle_radial = np.array(df['Radial_Angle'])
    X = []
    y = []
    for i in range(df.shape[0]-window):
        j = i + window
        loc_arr = emg_array[j-window:j]
        X.append(loc_arr)
        angles_flexion = all_angle_flexion[j-window:j]
        angles_radial = all_angle_radial[j-window:j]
        diff_flexion = difference(angles_flexion).mean()
        diff_radial = difference(angles_radial).mean()
        if(abs(diff_flexion)<angle_thresh and abs(diff_radial)<angle_thresh):
            y.append([1,0,0])
        elif(diff_flexion>0 and diff_radial>0):
            y.append([0,1,0])
        else:
            y.append([0,0,1])
    X = np.array(X)
    y = np.array(y)
        
    return X, y
```

```python
X, y = prep_data(df,300,'Pronation_Angle')
X.shape
```

```python
X, y = prep_data_DTM(df,300)
X.shape
```

## Multifile Stack

```python
files = []
path = 'Subjects/Fazil/day_04_24_21'
trial_names = ['Hammer','Cup'] # 'Bulb','Hammer' ,'Cup','Screw'
# trial_names = ['Bulb','Hammer' ,'Cup','Screw']
for exp in trial_names:
    for i in range(3):
        if(i==0):
            files.append(path+'/computed_'+exp+'.csv')
        else:
            files.append(path+'/computed_'+exp+'0'+str(i)+'.csv')
files
```

```python
files_df = []
for file in files:
    files_df.append(pd.read_csv(file,index_col=0))
```

```python
def multiple_prep_data(df_list,window,angle_label,Normalize=True):
    X_all, y_all = [], []
    for df in tqdm(df_list):
        X, y = prep_data(df,window,angle_label,Normalize)
        X_all.append(X)
        y_all.append(y)
        
    X_all_stack = np.vstack((X_all[0],X_all[1]))
    X_all_stack = np.vstack((X_all_stack,X_all[2]))
    
    y_all_stack = np.vstack((y_all[0],y_all[1]))
    y_all_stack = np.vstack((y_all_stack,y_all[2]))
    return X_all_stack, y_all_stack

def multiple_prep_data_DTM(df_list,window,Normalize=True):
    X_all, y_all = [], []
    for df in tqdm(df_list):
        X, y = prep_data_DTM(df,window,Normalize)
        X_all.append(X)
        y_all.append(y)
        
    X_all_stack = np.vstack((X_all[0],X_all[1]))
    X_all_stack = np.vstack((X_all_stack,X_all[2]))
    
    y_all_stack = np.vstack((y_all[0],y_all[1]))
    y_all_stack = np.vstack((y_all_stack,y_all[2]))
    return X_all_stack, y_all_stack
```

```python
X, y = multiple_prep_data(files_df,300,'Pronation_Angle',False)
```

```python
X, y = multiple_prep_data_DTM(files_df,300,False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
for data in [X_train, X_test, y_train, y_test]:
    print(data.shape)
```

```python
y_true = []
for d in y_test:
    idx = list(d).index(1)
    y_true.append(idx)
```

```python
def count(y):
    s0 = 0
    s1 = 0
    s2= 0
    for arr in y:
        if arr[0]==1:
            s0 = s0+1
        if arr[1]==1:
            s1 = s1+1
        if arr[2]==1:
            s2 = s2+1
    ls = [s0,s1,s2]
    for i in range(3):
        print(i,'-',ls[i])
```
